## Distribution plot/Histogram ---- univariable

Types
1. distplot -- univariant
2. jointplot -- bivariant --- many kind of plots
3. pairplot -- show all types of plots for all cols
4. rugplot -- bin size is 0
5. kdeplot -- only kde

```python
sns.distplot(df['total_bill']) 
```

kde = kernel density estimator in y axis
helps to find probability density function

```python
sns.displot(df['total_bill'], kde = False, bins = 20) #normal histogram
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'hex') #hexagon plot
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind = 'reg') #regression plot
```

```python
sns.pairplot(df)
#automatically pairs any 2 cols in a dataset.......... used only for min no, of columns
#only numerical
```

```python
sns.pairplot(df, hue = 'sex') #mention in hue for categorical variables
```
