# Plot title and Legends

```python
import numpy as np
```

```python
x_data = np.linspace(0,20)
y1_data = x_data**2
y2_data = x_data**3
y3_data = x_data**2.5

plt.plot(x_data,y1_data)
plt.plot(x_data,y2_data)
plt.plot(x_data,y3_data)
#pyplot will assign 3 different colours by default

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('3 lines on same graph')

plt.legend(['square','cube','mult by 2.5']) #not recommended method

plt.show()
```

```python
x_data = np.linspace(0,20)
y1_data = x_data**2
y2_data = x_data**3
y3_data = x_data**2.5

#specifying legend in plot (mostly preferred method)

plt.plot(x_data,y1_data, label = 'square')
plt.plot(x_data,y2_data, label = 'cube')
plt.plot(x_data,y3_data, label = 'multi of 2.5')
#pyplot will assign 3 different colours by default

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('3 lines on same graph')
#plt.legend(loc = 'best')#will place in best position
plt.legend(loc = 'center')

plt.show()
```

# Best version

```python
x_data = np.linspace(0,20)
y1_data = x_data**2
y2_data = x_data**3
y3_data = x_data**2.5

#specifying legend in plot (mostly preferred method)

plt.plot(x_data,y1_data,'--o', label = 'square')
plt.plot(x_data,y2_data, label = 'cube')
plt.plot(x_data,y3_data,':', linewidth = 5, label = 'multi of 2.5')
#pyplot will assign 3 different colours by default

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('3 lines on same graph')
#plt.legend(loc = 'best')#will place in best position
plt.legend(loc = 'center')

plt.show()
```
